```python
key = 'cell'

a = dataset_metadata_tables_csv[key]
b = dataset_metadata_tables_xlsx[key]
c = dataset_metadata_tables_db[key]

import numpy as np


id_ab_row = np.sum(a.values == b.values, axis=0) != a.shape[0]
id_ab_col = np.sum(a.values == b.values, axis=1) != a.shape[1]

id_ac_row = np.sum(a.values == c.values, axis=0) != a.shape[0]
id_ac_col = np.sum(a.values == c.values, axis=1) != a.shape[1]

id_bc_row = np.sum(b.values == c.values, axis=0) != b.shape[0]
id_bc_col = np.sum(b.values == c.values, axis=1) != b.shape[1]

print(np.allclose(a.iloc[id_ab_col,id_ab_row].replace('',-1),b.iloc[id_ab_col,id_ab_row].replace('',-1)))
print(np.allclose(a.iloc[id_ac_col,id_ac_row].replace('',-1),c.iloc[id_ac_col,id_ac_row].replace('NaN',-1)))
print(np.allclose(b.iloc[id_bc_col,id_bc_row].replace('',-1),c.iloc[id_bc_col,id_bc_row].replace('NaN',-1)))
```

The cell tables are different one to another yet close to a precision of relative ($10^{-5}$) and absolute ($10^{-8}$) tolerances.

#### The [README.txt](./data/hc-3/docs/hc3-original-docs/README.txt) file

> This directory contains the original metadata files which
describe the hc3 data set.  

>These original files should NOT
be used as the main source of information about the hc3 data set
because they have not been "cleaned up" and documented.

> Specifically, they contain extra information which is not
documented, make reference to some data which is not included
in the hc3 data set, and are harder to understand
than the metadata tables provided with the data set.

> These are provided in case the extra non-documented information
is useful and to document how the metadata tables provided
with the data set were generated.

> Included in this directory are the original data files:
KenjiData.mat and ElePosition.txt file.
